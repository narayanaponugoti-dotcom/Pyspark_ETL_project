```python
# Helper function to clean column names for Delta tables
import re

def clean_column_names(df):
    """Replace invalid characters in column names with underscores"""
    for col in df.columns:
        # Replace spaces and special characters with underscore
        new_col = re.sub(r'[^a-zA-Z0-9_]', '_', col)
        # Remove leading/trailing underscores and collapse multiple underscores
        new_col = re.sub(r'_+', '_', new_col).strip('_')
        if col != new_col:
            df = df.withColumnRenamed(col, new_col)
    return df

print("Helper function defined successfully")
```

```python
# Create table for customers data
df = spark.read.csv(
    "/Volumes/my_catalog/bronze/etl_project1/customers-100000.csv",
    header=True,
    inferSchema=True
)
df_clean = clean_column_names(df)
df_clean.write.mode("overwrite").saveAsTable("my_catalog.bronze.customers")

print("✓ Created table: my_catalog.bronze.customers")
```
