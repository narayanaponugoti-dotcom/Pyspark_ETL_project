```python
# Load DataFrame from the gold layer transformation pipeline
gold_df = spark.table("my_catalog.gold.product_performance")

# Generate summary statistics for dashboard and analytics
summary_df = gold_df.groupBy("Category").agg(
    {"Price": "sum", "Product_Id": "count"}
).withColumnRenamed("sum(Price)", "total_amount") \
 .withColumnRenamed("count(Product_Id)", "record_count")

# Display summary for dashboard integration
display(summary_df)
```

```python
# Create dashboard widgets for interactive filtering
categories = [row[0] for row in gold_df.select("Category").distinct().collect()]
dbutils.widgets.dropdown("Category", "All", [str(c) for c in categories] + ["All"])

# Filter DataFrame based on widget selection
selected_category = dbutils.widgets.get("Category")
filtered_df = gold_df if selected_category == "All" else gold_df.filter(gold_df.Category == selected_category)

# Generate summary statistics for dashboard
dashboard_df = filtered_df.groupBy("Category").agg(
    {"Price": "sum", "Product_Id": "count"}
).withColumnRenamed("sum(Price)", "total_amount") \
 .withColumnRenamed("count(Product_Id)", "record_count")

# Display dashboard summary
display(dashboard_df)
```
